```python
# Define a list to store shopping items
shopping_list = []

# Dictionary to store item categories
item_categories = {"fruits": set(), "veggies": set(), "dairy": set(), "utensils": set(), "others": set()}

# Function to display the menu
def display_menu():
    print("\nShopping List Manager")
    print("1. Add item")
    print("2. Remove item")
    print("3. View list")
    print("4. Proceed to Order")

while True:
    display_menu()
    choice = input("Enter your choice (1, 2, 3, or 4): ")
    
    if choice == "1":
        remaining_slots = 10 - len(shopping_list)
        if remaining_slots > 0:
            num_items = int(input(f"You can add up to {remaining_slots} items. How many would you like to add? "))
            num_items = min(num_items, remaining_slots) 
            
            for _ in range(num_items):
                item = input("Enter item to add:  ")
                category = input("Enter category (fruits, vegetables, utensils, dairy, others): ").lower()
                while category not in item_categories:
                    category = input("Invalid category. Please enter one of (fruits, vegetables, dairy, others): ").lower()
                shopping_list.append(item)
                item_categories[category].add(item)
                print(f"{item} added to shopping list under {category} category.")
        else:
            print("Shopping list is full! You cannot add more than 10 items.")
    
    elif choice == "2":
        if shopping_list:
            item = input("Enter item to remove (exact name as in list): ")
            if item in shopping_list:
                shopping_list.remove(item)
                for category in item_categories:
                    if item in item_categories[category]:
                        item_categories[category].remove(item)
                        break
                print(f"{item} removed from shopping list.")
            else:
                print("Item not found in shopping list. Please check the spelling.")
        else:
            print("Shopping list is empty. Nothing to remove.")
    
    elif choice == "3":
        if shopping_list:
            print("\nYour Shopping List:")
            for index, item in enumerate(shopping_list, start=1):
                print(f"{index}. {item}")
            print("\nCategories:")
            for category, items in item_categories.items():
                print(f"{category.capitalize()}: {', '.join(items) if items else 'None'}")
        else:
            print("Your shopping list is empty.")
    
    elif choice == "4":
        print("Your order has been processed. Thank you for shopping with us!")
        break
    
    else:
        print("Invalid choice. Please enter 1, 2, 3, or 4.")
```
